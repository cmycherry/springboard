```python
x = [i**2 for i in range(100)]
```

```python
for y in x:
    print(y)
```

```python
for y in enumerate(x):
    print(y)
```

```python
for i,y in enumerate(x):
    print(i)
    print(y)
```

```python
z = list(enumerate(x))
print(z)
```

```python
i,x = zip(*z)
print(i)
print(x)
```

```python
z=zip(i,x)
print(z)
```

```python
print(list(z))
```
