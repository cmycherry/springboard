```python
#read the data file of Dow-Jones historical data
import pandas as pd
df = pd.read_csv('dowjoneshistoricalprices.csv')
print(df.columns)

#Clear the data formate of columns name by low
df.columns = [str.strip(name) for name in df.columns]
df.columns = [str.lower(name) for name in df.columns]
print(df.columns)

#insert columns of year, month, and quarters into the data frame
df.index = pd.to_datetime(df.date)
df['year'] = df.index.year
df['month'] = df.index.month
df['quarter'] = df.index.quarter
df['day of the week']=df.index.weekday
df['changes']= df['close']-df['open']
print(df)

# describe function: Generates descriptive statistics that summarize the central tendency, dispersion and shape of a datasetâ€™s distribution, excluding NaN values.
df['changes'].describe()
df.describe()

```

```python
predictions = sgd.predict(df[['quarter','month']])
```
